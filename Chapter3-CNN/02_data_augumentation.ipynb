## 数据增强    
是对抗过拟合问题的一个重要方法   

常用方法：
- flip
- rotation
- resize
- crop
- scale
- add noise

```
import sys
sys.path.append('..')
from PIL import Image
from torchvision import transforms 
```

```
!mkdir img/
!mv lena.png img/
```

```
img = Image.open('./img/lena.png')
img
```

```
# flip
h_flip = transforms.RandomHorizontalFlip()(img)
h_flip
```

```
v = transforms.RandomVerticalFlip()(img)
v
```

```
# scale
scale_img = transforms.Resize((100, 100))(img)
scale_img
```

```
img.size, scale_img.size
```

```
# crop
crop1 = transforms.RandomCrop((100, 100))(img)
crop1
```

```
crop2 = transforms.CenterCrop((100, 100))(img)
crop2
```

```
# rotation
rot = transforms.RandomRotation(90)(img)
rot
```

```
# bright
# 随机从 0 ~ 2 之间亮度变化，1 表示原图
bright_img = transforms.ColorJitter(brightness=1.2)(img)
bright_img
```

```
# contrast
# 随机从 0 ~ 2 之间对比度变化，1 表示原图
contrast_img = transforms.ColorJitter(1.2)(img)
contrast_img
```

```
# color
# 随机从 -0.5 ~ 0.5 之间对颜色变化
color_img = transforms.ColorJitter(hue=0.2)(img) 
color_img
```
